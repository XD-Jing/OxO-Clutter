```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
from scipy.stats import norm
%matplotlib inline
```

```python
import pandas as pd
um = pd.read_excel('./sMDT-Tube-Information.xlsx', sheet_name = 'UM')
mpi = pd.read_excel('./sMDT-Tube-Information.xlsx', sheet_name = 'MPI')
```

```python
print(um.columns)
print(mpi.columns)
```

```python
binX = np.linspace(1128.6, 1129.7, 22)
dX = binX[2] - binX[1]
fig = plt.figure()
length_mpi = mpi['Length(mm)'].values
length_um = um['Length-AfterSwagging(mm)'].values


length_mpi = length_mpi[~np.isnan(length_mpi)]
length_um = length_um[~np.isnan(length_um)]


alpha = 1
plt.hist( length_mpi, bins = binX, label = 'MPI', alpha = alpha)
plt.hist( length_um, bins = binX, label = 'UM', alpha = alpha)


length_mpi1 = length_mpi[length_mpi<1129.05]
binX1 = np.linspace(1128.6, 1129.05, 30)
(mu_mpi1, sigma_mpi1) = norm.fit(length_mpi1)
fit_mpi1 = mlab.normpdf( binX1, mu_mpi1, sigma_mpi1) 
l = plt.plot(binX1, fit_mpi1 * len(length_mpi1) * dX, 'k--', linewidth = 2)


length_mpi2 = length_mpi[length_mpi>1129.05]
binX2 = np.linspace(1129.05, 1129.6, 50)
(mu_mpi2, sigma_mpi2) = norm.fit(length_mpi2)
fit_mpi2 = mlab.normpdf( binX2, mu_mpi2, sigma_mpi2) 
l = plt.plot(binX2, fit_mpi2 * len(length_mpi2) * dX, 'k--', linewidth = 2)


binX3 = np.linspace(1128.9, 1129.6, 50)
(mu_um, sigma_um) = norm.fit(length_um)
fit_um = mlab.normpdf( binX3, mu_um, sigma_um) 
l = plt.plot(binX3, fit_um * len(length_um) * dX, 'k--', linewidth = 2, label = 'Fit')


plt.text(x = 1128.65, y = 65, s = 'sMDT', fontsize = 18)
plt.text(x = 1128.65, y = 55, s = 'UM: {} tubes'.format(len(length_um)), fontsize = 12)
plt.text(x = 1128.75, y = 50, s = r'{:.2f}$\pm${:.2f}mm'.format(mu_um, sigma_um), fontsize = 12)
plt.text(x = 1128.65, y = 45, s = 'MPI: {} tubes'.format(len(length_mpi)), fontsize = 12)
plt.text(x = 1128.75, y = 40, s = r'{:.2f}$\pm${:.2f}mm'.format(mu_mpi1, sigma_mpi1), fontsize = 12)
plt.text(x = 1128.75, y = 35, s = r'{:.2f}$\pm${:.2f}mm'.format(mu_mpi2, sigma_mpi2), fontsize = 12)


plt.xlabel('Length[mm]')
plt.ylabel('Tubes')
plt.legend()
plt.savefig('length.png', dpi=200)
plt.savefig('length.eps')
```

```python
binX = np.linspace(335, 380, 9)
dX = binX[2] - binX[1]
fig = plt.figure()
tension_mpi = mpi['Database-Tension(g)'].values
tension_mpi_um = mpi['Tension(g)'].values
tension_um = um['Tension-OneWeek(g)'].values[7:]
# tension_um = um['Tension-OneWeek(g)'].values

tension_mpi = tension_mpi[~np.isnan(tension_mpi)]
tension_mpi_um = tension_mpi_um[~np.isnan(tension_mpi_um)]
tension_um = tension_um[~np.isnan(tension_um)]


def drawText(x, y, tension, text):
    plt.text(x = x , y = y, s = text, fontsize = 12)
    plt.text(x = x + 2 , y = y-15, s = r'{} tubes: {:.2f}$\pm${:.2f}g'.format(len(tension), np.mean(tension), np.std(tension)), fontsize = 12)

x, y = 335, 180
plt.text(x = x + 36 , y = y, s = 'sMDT', fontsize = 18)
drawText(x = x , y = y-40, tension = tension_um, text = 'UM @ UM')
drawText(x = x , y = y-70, tension = tension_mpi, text = 'MPI @ MPI')
drawText(x = x , y = y-100, tension = tension_mpi_um, text = 'MPI @ UM')


alpha = 0.5
plt.hist( tension_mpi, bins = binX, label = 'MPI Database', alpha = alpha)
plt.hist( tension_mpi_um, bins = binX, label = 'MPI by UM', alpha = alpha)
plt.hist( tension_um, bins = binX, label = 'UM', alpha = alpha)
plt.xlabel('Tension[g]')
plt.ylabel('Tubes')
plt.legend()
plt.savefig('tension.png', dpi=200)
plt.savefig('tension.eps')
```
